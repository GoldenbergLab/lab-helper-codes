### The purpose of this code is to get user information from twitter api. The input is a txt file with user names, the output is all of the user's information. 


```python
import json
import tweepy
import pandas as pd

#add personal keys
ckey = #add
csecret = #add
atoken = #add
asecret = #add


auth = tweepy.OAuthHandler(ckey, csecret)
auth.set_access_token(atoken, asecret)

api = tweepy.API(auth)

#change the name of file
fin = open ('nameOfFile.txt', "r")
#fin = open(sys.argv[1], "rt")
querylist = list()
allusersdata = list()

#loop through the line
for line in fin: # for each line 
    userId = line.replace("\n","") #append until it gets to 100 (while removing new line) 
    querylist.append (userId)
    print (userId)
    try:
        usersdata = api.get_user(userId) # get the value 
    except tweepy.TweepError as e:
        print (e)
        usersdata = e
    allusersdata.append(usersdata) #all users information in a long list allusersdata

        
#organize the information you need
d = []
for ud in allusersdata:
    try:
        #print (ud.screen_name + "\t" + str(ud.id_str))
        name_id = [ud.screen_name,ud.id_str,ud.followers_count] #you can add whatever you want here instead
    except:
        #print ("Couldn't find") 
        name_id = ["no_name", "no_user", "no_followers"]
    d.append(name_id)
    
d= pd.DataFrame.from_dict(d)
d.columns=['name','userNumber','followers'] #change collumn names in accordance to whatever you need. 
d['userNames'] = querylist #add users 
```
